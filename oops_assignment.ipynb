#4 How does Python implement method overloading? Give an example.

Ans : 
Python does not support method overloading like other programming languages such as Java or C++. This means we cannot define multiple methods with the same name but different arguments in a class. If we try to do so, the latest method definition will overwrite the previous ones.

How Does Python Handle Method Overloading?

Python allows you to simulate method overloading using:
1. Default Arguments: Provide default values for parameters.
2. Variable-Length Arguments: Use *args for any number of arguments.
3. Type Checking: Check argument types inside the method and decide behavior.

Example 1: Using Default Arguments


class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(10))          # Output: 10
print(calc.add(10, 20))      # Output: 30
print(calc.add(10, 20, 30))  # Output: 60



Example 2: Using *args


class Calculator:
    def add(self, *args):
        return sum(args)

calc = Calculator()
print(calc.add(10, 20))           # Output: 30
print(calc.add(10, 20, 30, 40))   # Output: 100

Example 3: Type Checking


class Calculator:
    def add(self, a, b):
        if isinstance(a, str) or isinstance(b, str):
            return str(a) + str(b)  # Concatenate strings
        return a + b  # Add numbers

calc = Calculator()
print(calc.add(10, 20))           # Output: 30
print(calc.add("Hello, ", "World!"))  # Output: Hello, World!

#6 Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.

Ans :
In Python, inheritance allows a class to acquire properties and methods from another class. There are five types of inheritance:

1. Single Inheritance  
   In this type, a child class inherits from a single parent class.  
   Example: A class Dog inherits from a class Animal.

2. Multiple Inheritance  
   In this type, a child class inherits from more than one parent class.  
   Example: A class Student inherits from both Person and Learner.

3. Multilevel Inheritance  
   In this type, a class inherits from another class, which itself is inherited from another class.  
   Example: A class Grandchild inherits from Child, which inherits from Parent.

4. Hierarchical Inheritance  
   In this type, multiple child classes inherit from the same parent class.  
   Example: Classes Cat and Dog inherit from the class Animal.

5. Hybrid Inheritance  
   This is a combination of two or more types of inheritance, such as a mix of multiple and hierarchical inheritance.  
   Example: A complex structure combining features of different inheritance types.

Example of Multiple Inheritance:


class Person:
    def speak(self):
        return "I am a person"

class Worker:
    def work(self):
        return "I am working"

class Student(Person, Worker):
    def study(self):
        return "I am studying"

# Usage
s = Student()
print(s.speak())  # Output: I am a person
print(s.work())   # Output: I am working
print(s.study())  # Output: I am studying
```

In this example, the Student class inherits methods from both Person and Worker classes, demonstrating multiple inheritance.

#7 What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

Ans : 
Method Resolution Order (MRO) in Python is the order in which Python searches for a method or attribute in a class hierarchy. It is used in the context of inheritance to determine the sequence of classes that are checked when a method is called.

MRO follows the C3 linearization algorithm, which ensures a consistent and predictable order. It searches the current class first, then moves to its parent classes, and so on. If multiple inheritance is involved, it ensures that the method is only called once per class in the hierarchy.

You can retrieve the MRO programmatically using the mro method or the _mro_ attribute.

Example:


class A:
    pass

class B(A):
    pass

class C(B):
    pass

# Retrieve MRO using mro method
print(C.mro())

# Retrieve MRO using _mro_ attribute
print(C._mro_)

#8 Create an abstract base class Shape with an abstract method area(). Then create two subclasses
#Circle and Rectangle that implement the area() method.
Ans : 
Here is an example of an abstract base class Shape with an abstract method area and its subclasses Circle and Rectangle that implement the area method:

python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Example usage
circle = Circle(5)
print("Circle Area:", circle.area())

rectangle = Rectangle(4, 6)
print("Rectangle Area:", rectangle.area())


Explanation:

1. Shape is an abstract base class that inherits from ABC (Abstract Base Class) and defines the abstract method area.
2. Circle and Rectangle are subclasses of Shape that implement the area method with their specific calculations.
3. The area method in Circle calculates the area using the formula for the area of a circle.
4. The area method in Rectangle calculates the area using the formula for the area of a rectangle.

Output:
Circle Area: 78.5  
Rectangle Area: 24
```

````python
#11 Write a class that overrides the __str__ and __add__ magic methods. What will these methods allow
#you to do?
Ans : 
class Person:
    def _init_(self, name, age):
        self.name = name
        self.age = age

    def _str_(self):
        return f"Person(name: {self.name}, age: {self.age})"

    def _add_(self, other):
        if isinstance(other, Person):
            return self.age + other.age
        return "Cannot add different types"

# Example usage
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

# Using _str_ to print the object in a custom way
print(person1)  # Output: Person(name: Alice, age: 30)

# Using _add_ to add ages of two Person objects
print(person1 + person2)  # Output: 55
````

#12 Create a decorator that measures and prints the execution time of a function.

import time

def measure_execution_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Execution time of {func._name_}: {execution_time} seconds")
        return result
    return wrapper

# Example usage
@measure_execution_time
def some_function():
    total = 0
    for i in range(1000000):
        total += i
    return total

#13 Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

Ans : 
The Diamond Problem arises in multiple inheritance when a class inherits from two classes that both inherit from a common parent class. This creates a diamond-shaped inheritance structure. The issue occurs when the child class ends up inheriting methods or properties from the parent class through both its parents, leading to ambiguity about which method or property should be used.

To better understand the problem, imagine the following structure:


       A
      / \
     B   C
      \ /
       D


In this scenario, class D inherits from both B and C, while both B and C inherit from A. If both B and C override methods from A, and D inherits from both, there could be confusion about which version of the method D should inherit.

Python handles this problem using the C3 Linearization algorithm, which is designed to provide a consistent method resolution order (MRO). The MRO ensures that each class in the inheritance chain is considered only once, and it specifies the order in which classes are searched for methods.

Here's an example that demonstrates how Python handles the Diamond Problem:


class A:
    def hello(self):
        print("Hello from A")

class B(A):
    def hello(self):
        print("Hello from B")

class C(A):
    def hello(self):
        print("Hello from C")

class D(B, C):
    pass

d = D()
d.hello()


In this example, classes B and C both override the hello method of class A. Class D inherits from both B and C. When the hello() method is called on an object of class D, Python uses the MRO to determine which method to execute. According to Pythonâ€™s MRO, class B is checked first, then class C, and finally class A.

The output of this code will be:


Hello from B


Python resolves this by using the C3 Linearization, which ensures that each class is called in a clear, predictable order. You can see the MRO of class D by calling D.mro(), which returns the order in which methods will be resolved:

python
print(D.mro())


This will output the MRO, showing the classes Python checks in order when looking for methods.

In conclusion, Python solves the Diamond Problem by using the C3 Linearization method, ensuring that the classes are checked in a consistent and clear order, avoiding ambiguity.

#15 Implement a static method in a class that checks if a given year is a leap year.

class YearChecker:
    
    @staticmethod
    def is_leap_year(year):
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        return False

# Example usage
year = 2024
if YearChecker.is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")


Explanation:

1. The class YearChecker contains a static method is_leap_year, which takes a year as an argument and checks if it is a leap year.
2. A year is a leap year if:
   - It is divisible by 4, but not divisible by 100, or
   - It is divisible by 400.
3. The static method is defined using the @staticmethod decorator, meaning it does not require an instance of the class to be called.
4. The method returns True if the year is a leap year, and False otherwise.
5. In the example usage, the method is called with the year 2024, and it prints whether the year is a leap year or not.

Output:


2024 is a leap year.
